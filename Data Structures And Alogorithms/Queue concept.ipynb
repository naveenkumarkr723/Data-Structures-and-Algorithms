## Queue Implementation Using List

### for enqueue we used the appending method
### for dequeue we used the pop method



```python
## this method add the elements from the right 

queue = []

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)

queue
```

```python
### this queue.pop(0) will remove the elements what we add at first in queue

queue.pop(0)
queue
```

```python
## this method add from the left

queue = []
queue.insert(0,10)
queue.insert(0,20)
queue.insert(0,30)

queue
```

```python
### this queue.pop() will remove the elements what we add at first in queue from the left

queue.pop()
queue
```

```python
### Example of the stack 

stack = []

## defining the push operation
def pushing():
    push = int(input("enter the element which you are want to push"))
    stack.append(push)
    print(stack)
    
def removing():
    if not stack:
        print("the stack is empty ")
    else:
        #print("enter the element which you are want to pop or remove from the last ")
        e = stack.pop(0)
        
        print("the removed element is the :",e)
        print(stack)
    
    
    
while True:
    print("select the options 1.push 2.pop 3.quit the program")
    choice = int(input())
    
    if choice == 1:
        pushing()
    elif choice == 2:
        removing()
    elif choice == 3:
        print("the stack after performing the tasks :",stack)
        break
    else:
        print("enter the correct options")
```

```python
## 1. Collection --->  deque(double ended queue)

import collections 
stack = collections.deque()
stack

stack.appendleft(10)
stack.appendleft(20)
stack.appendleft(30)

## for the appendleft we use the pop

stack.pop()
stack.pop()
stack.pop()

#stack.pop()
stack
```
