```python
## Traversal method of the LinkedList

class Node:

    def __init__(self, data):
        self.data = data
        self.nref = None
        self.pref = None


class LinkedList:
    
    def __init__(self):
        self.head = None

    def print_LL_forward (self):
        if self.head is None:
            print("inked list is empty! , to print in forward direction")
        else:
            n = self.head
            while n is not None: 
                print(n.data)
                n = n.nref
    
    def print_LL_backward (self):
        if self.head is None:
            print("inked list is empty! ,to print in backward direction")
        
        else:
            n = self.head
            
            while n.nref is not None: 
                n = n.nref
            
            while n is not None: 
                print(n.data)
                n = n.pref              

    def insert_empty(self,data):
        if self.head is None:
            
            new_node = Node(data)
            self.head = new_node
        else:
            print("Linked List is not empty")
            
            
    def add_begin(self, data):
        new_node = Node(data)
       
        if self.head is None:
            self.head  = new_node
        else:
            new_node.nref  = self.head
            self.head.pref = new_node
            self.head      = new_node
   


    def add_end(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
        else:
            n  = self.head
            while n is not None:
                n =  n.nref
            n.nref = new_node                
            new_node.pref = n
      
    
    ### addition of new nodes in between the nodes
    
    def add_after(self,data,x):
        if self.head is None:
            print("the double linked list is empty in add after method")
        else:
            n  = self.head
            while n is not None:
                if x == n.data :
                    break
                n = n.nref
                
            if n is None:
                print("the given node is not present in the add_after method") 
            else:
                new_node = Node(data)
                new_node.nref = n.nref
                new_node.pref = n
                if n.nref is not None:
                    n.nref.pref = new_node
                    
                n.nref = new_node
                
                
    def add_before(self,data,x):
        if self.head is None:
            print("the double linked list is empty in add after method")
        else:
            n  = self.head
            while n.nref is not None:
                if x == n.data :   
                    break
                n = n.nref
                
            if n is None:
                print("the given node is not present in the add_after method")
                          
            else:
                new_node = Node(data)
                new_node.nref = n
                new_node.pref = n.pref
                if n.pref is not None:
                    n.pref.nref = new_node
                else:  
                    self.head = new_node
                n.pref = new_node
                
                
    def delete_begin(self):
        if self.head is None:
            print("the double linked list is empty can't delete by delete_begin method")
            return
        if self.head.nref is None:
            self.head = None
            print("deleted the given first node , only one node is present in the given doubly linked list")
            
        else:
            self.head = self.head.nref
            self.head.pref = None
    
    
    def delete_end(self):
        if self.head is None:
            print("the double linked list is empty can't delete by delete_end method")
            return
        if self.head.nref is None:
            self.head = None
            print("deleted the given first node , only one node is present in the given doubly linked list")
        else:   
            n  = self.head
            while n.nref is not None:
                n = n.nref  
            n.pref.nref = None        
                    
    
    
    def delete_by_value(self, x):
        if self.head is None:
            print("the double linked list is empty can't delete by delete_end method")
            return
        
        if self.head.nref is None:
            if self.head.data == x : 
                self.head = None
            else:
                print("the given x value not present in the  first node")

            return
        if self.head.data == x :
            self.head = self.head.nref
            self.head.pref = None
            return
        n = self.head
        
        while n.nref is not None:
            if x == n.data :
                break
            n= n.nref
        if n.nref is not None:
            n.nref.pref = n.pref
            n.pref.nref = n.nref
            
        else:
            if n.data == x:
                n.pref.nref = None
            else :
                print("x is not present in the given dll !")
    
    
LL1 = LinkedList()
LL1.add_begin(115)
LL1.add_begin(15)
LL1.add_begin(895)


LL1.print_LL_forward()
```
