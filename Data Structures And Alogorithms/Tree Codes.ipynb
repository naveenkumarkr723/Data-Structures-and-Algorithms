```python
### binary search tree first code

class BST():
    
    def __init__(self , key):
        self.key = key
        self.lchild  = None
        self.rchild  = None

        
binary = BST(10) 
print("the root node of the tree : ",binary.key)
print(binary.lchild)
print(binary.rchild)


###  adding the left child to the root node 
binary.lchild = BST(7) 
print("the left child of the root node of the tree : ",binary.lchild.key)
print(binary.lchild.lchild)
print(binary.lchild.rchild)


###  adding the right child to the root node 
binary.rchild = BST(17) 
print("the right child of the root node of the tree : ",binary.rchild.key)
print(binary.rchild.lchild)
print(binary.rchild.rchild)
```

```python
### creating left and right childs

class BST():
    
    def __init__(self , key):
        self.key = key
        self.lchild  = None
        self.rchild  = None

    def insert(self,data):
        if self.key is None:
            self.key = data
            return 
        
        if self.key == data:
            return
            
        
        if self.key > data:
            if self.lchild :
                self.lchild.insert(data)
            else:
                self.lchild = BST(data)
        else:
            if self.rchild :
                self.rchild.insert(data)
            else:
                self.rchild = BST(data)
            
     
    def search(self,data):
        if self.key == data:
            print("Node has found!")
            return
        if self.key > data:
            if self.lchild :
                self.lchild.search(data)
            else:
                print("Node is not found in left subtree")
        else:
            if self.rchild :
                self.rchild.search(data)
            else:
                print("Node is not found in right subtree")
            
    def preorder(self):
        print(self.key,end = " ")
        if self.lchild :
            self.lchild.preorder()
        if self.rchild :
            self.rchild.preorder()
        
    def inorder(self):
        if self.lchild :
            self.lchild.inorder()
        print(self.key,end = " ")
        if self.rchild :
            self.rchild.inorder()
            
            
    def postorder(self):
        if self.lchild :
            self.lchild.postorder()
        if self.rchild :
            self.rchild.postorder()
        print(self.key,end = " ")
        
    def delete(self,data):
        if self.key is None:
            print("Tree is Empty!")
            return
        if self.key > data:
            if self.lchild :
                self.lchild = self.lchild.delete(data)
            else:
                print("the given Node is Not present")
        elif self.key < data:
            if self.rchild :
                self.rchild = self.rchild.delete(data)
            else:
                print("the given Node is Not present")
            
        else:
            if self.lchild is None:
                temp = self.rchild
                self = None 
                return temp
            if self.rchild is None:
                temp = self.lchild
                
                self = None 
                return temp
            node = self.rchild
            while node.lchild:
                node = node.lchild
            self.key = node.key
            self.rchild = self.rchild.delete(node.key)
        return self
        
    def min_node(self):
        current = self
        while current.lchild:
            current = current.lchild
        print("the minimum key node in tree is :  ",current.key)
        
    def max_node(self):
        current = self
        while current.rchild:
            current = current.rchild
        print("the minimum key node in tree is :  ",current.key)
        

        
        
root = BST(10)     #root = BST(None) --> it will create the root node as none       
#root.insert(20) 
#root.insert(40) 
#root.insert(5) 
            
list1 = [2,3,4,5,6,10,2,45,52,32]            
for i in list1:
    root.insert(i)
#root.search(10) 

print("the preorder of the tree is found! sequence it follows as root left right ",root.preorder())
#print("the Inorder of the tree is found! sequence it follows as left root right ",root.inorder())
#print("the postorder of the tree is found! sequence it follows as left right root  ",root.postorder())

#root.delete(2)
#root.preorder()

root.max_node()
root.min_node()
```
